```python
def bar_chart(feature):
    survived = ttrain[ttrain['Survived']==1][feature].value_counts()
    dead = ttrain[ttrain['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index =['Survived','Dead']
    df.plot(kind='bar',stacked=True)
```

```python
# filling missing value 
Pclass1 = ttrain[ttrain['Pclass']==1]['Embarked'].value_counts()
Pclass2 = ttrain[ttrain['Pclass']==2]['Embarked'].value_counts()
Pclass3 = ttrain[ttrain['Pclass']==3]['Embarked'].value_counts()
df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index = ['1st class','2nd class', '3rd class']
df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
# fill missing Fare with median fare for each Pclass
ttrain["Fare"].fillna(ttrain.groupby("Pclass")["Fare"].transform("median"), inplace=True)
ttest["Fare"].fillna(ttest.groupby("Pclass")["Fare"].transform("median"), inplace=True)
ttrain.head(50)
```

```python
for dataset in traintest_dta:
    dataset.loc[ dataset['Fare'] <= 17, 'Fare'] = 0,
    dataset.loc[(dataset['Fare'] > 17) & (dataset['Fare'] <= 30), 'Fare'] = 1,
    dataset.loc[(dataset['Fare'] > 30) & (dataset['Fare'] <= 100), 'Fare'] = 2,
    dataset.loc[ dataset['Fare'] > 100, 'Fare'] = 3
```

```python
Pclass1 = ttrain[ttrain['Pclass']==1]['Cabin'].value_counts()
Pclass2 = ttrain[ttrain['Pclass']==2]['Cabin'].value_counts()
Pclass3 = ttrain[ttrain['Pclass']==3]['Cabin'].value_counts()
df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index = ['1st class','2nd class', '3rd class']
df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
cabin_mapping = {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}
for dataset in traintest_dta:
    dataset['Cabin'] = dataset['Cabin'].map(cabin_mapping)
```

```python
# fill missing Fare with median fare for each Pclass
ttrain["Cabin"].fillna(ttrain.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
ttest["Cabin"].fillna(ttest.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
```

```python
ttrain["FamilySize"] = ttrain["SibSp"] + ttrain["Parch"] + 1
ttest["FamilySize"] = ttest["SibSp"] + ttest["Parch"] + 1
```

```python
family_mapping = {1: 0, 2: 0.4, 3: 0.8, 4: 1.2, 5: 1.6, 6: 2, 7: 2.4, 8: 2.8, 9: 3.2, 10: 3.6, 11: 4}
for dataset in traintest_dta:
    dataset['FamilySize'] = dataset['FamilySize'].map(family_mapping)
```

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors = 7)
scoring = 'accuracy'
score = cross_val_score(clf, ttrain_dta, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# kNN Score
round(np.mean(score)*100, 2)
```
